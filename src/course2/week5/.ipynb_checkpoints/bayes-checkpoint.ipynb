```python
from sklearn.cross_validation import cross_val_score
import sklearn.naive_bayes as naive_bayes
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import sklearn.datasets as datasets
matplotlib.style.use('ggplot')
%matplotlib inline
```

```python
digits = datasets.load_digits()
cancer = datasets.load_breast_cancer()
```

```python
bernoulli = naive_bayes.BernoulliNB()
multinomial = naive_bayes.MultinomialNB()
gaussian = naive_bayes.GaussianNB()
```

```python
cv_score = cross_val_score(bernoulli, digits.data, digits.target, cv  = 10).mean()
print(cv_score)
cv_score = cross_val_score(multinomial, digits.data, digits.target, cv  = 10).mean()
print(cv_score)
cv_score = cross_val_score(gaussian, digits.data, digits.target, cv  = 10).mean()
print(cv_score)
```

```python
cv_score = cross_val_score(bernoulli, cancer.data, cancer.target, cv  = 10).mean()
print(cv_score)
cv_score = cross_val_score(multinomial, cancer.data, cancer.target, cv  = 10).mean()
print(cv_score)
cv_score = cross_val_score(gaussian, cancer.data, cancer.target, cv  = 10).mean()
print(cv_score)
```
